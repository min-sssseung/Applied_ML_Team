```python
from sklearn.preprocessing import StandardScaler
scaler= StandardScaler()
df4 = df3.copy()

saved_df = df4.iloc[:, :5]

cols = df4.columns[5:]
scaled_df = pd.DataFrame(scaler.fit_transform(df4.iloc[:, 5:]), columns=cols)

df5 = pd.concat([saved_df, scaled_df], axis=1)
df5['ESG'] = pd.Categorical(df5['ESG'], categories=['A+', 'A', 'B+', 'B', 'C', 'D'], ordered=True)
sns.countplot(data=df5, x='ESG') # dataset imbalance
```

```python
sns.pairplot(df5, hue='ESG')
```

```python
import numpy as np

correlation_matrix = df5.iloc[:, 5:].corr()

# Create a mask for the upper triangle
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))

# Set up the matplotlib figure
fig, ax = plt.subplots(figsize=(12, 10))

# Draw the heatmap with the mask
sns.heatmap(correlation_matrix, mask=mask, cmap="coolwarm", vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})

plt.title('Correlation Plot')
plt.show()
```

```python
from bs4 import BeautifulSoup
import pandas as pd

corp_code = []
for i in range(1,106):
    url = f'https://www.cgs.or.kr/business/esg_tab04.jsp?pg={i}&pp=10&skey=&svalue=&sfyear=2023&styear=2023&sgtype=&sgrade=#ui_contents;'
    df = pd.read_html(url)
    corp_code += list(df[1].iloc[:,2])
```

```python
corp_code = [str(code) for code in corp_code]

def add_zero(c):
    if len(c) < 6:
        c = (6-len(c)) * '0' + c
    return c

corp_code = list(map(add_zero,corp_code))
    
```

```python
with open('corp_code.txt','w') as file:
    for c in corp_code:
        file.write(str(c)+'\n')
```

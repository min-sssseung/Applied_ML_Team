```python
df1 = pd.read_csv('C:/Users/kmk45/OneDrive/바탕 화면/AML/Main/Financial/ESG_final.csv', dtype={'corp_code': str, 'stock_code': str})
# df1 = df1.drop(['Unnamed: 0'], axis=1)
df1.head()
```

```python
df1['시가총액'] = df1.apply(lambda x:stock.get_market_cap(x['연도']+'0101', x['연도']+'1231', x['stock_code']).iloc[:, 0].mean(), axis=1)
```

```python
values = []
for code, year in df1[['stock_code', '연도']].values:
    start, end = year + '0101', year + '1231'
    v = stock.get_market_cap(start, end, code)
    values.append(v.iloc[:, 0].mean())
    print(code)

# df1['시가총액'] = values   
```

```python
df1['시가총액'] = values   

df1.head()
```
